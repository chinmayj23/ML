```
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler

from sklearn import linear_model
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

np.random.seed(27)
```

```
data=pd.read_csv("/content/drive/My Drive/kc2.csv")
```

```
data.head()
```

```
data.info()
```

```
data.loc[0:19]
```

```
data.loc[:,:'branchCount']
```

```
scaler= StandardScaler()
df = scaler.fit_transform(data.loc[:,:'branchCount'])
```

```
y=data['problems']
```

```
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(y.unique())
y = le.transform(y)
```

```
seed = 8
test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=test_size, random_state=seed)
```

```
logistic = linear_model.LogisticRegression()
sgd = linear_model.SGDClassifier()
```

```
scores = cross_val_score(logistic, X_train, y_train, cv=5, scoring='roc_auc')
np.mean(scores)
np.std(scores)
```

```
scores = cross_val_score(sgd, X_train, y_train, cv=5, scoring='roc_auc')
np.mean(scores)
np.std(scores)
```

```
penalty = ['l1', 'l2']
C = [0.0001, 0.001, 0.01]
class_weight = [{1:0.5, 0:0.5}, {1:0.6, 0:0.4}]
solver = ['liblinear', 'saga'] #

paramgrid = dict(penalty=penalty, C=C, class_weight=class_weight, solver=solver)

grid = GridSearchCV(estimator=logistic, param_grid=paramgrid, scoring='roc_auc', verbose=1, n_jobs=-1)
grid_result = grid.fit(X_train, y_train)
```

```
grid_result.best_score_
grid_result.best_params_
```

```
y_test_pred = grid_result.predict(X_test)
y_test_pred
```

```
from sklearn.metrics import accuracy_score
ans = accuracy_score(y_test, y_test_pred)
```

```
ans
```

```
penalty = ['l1', 'l2']
eta0 = [0.5, 1, 1.5]
loss= ['hinge','log']
alpha = [0.0001, 0.001]
learning_rate = ['constant', 'optimal']
class_weight = [{1:0.5, 0:0.5}, {1:0.6, 0:0.4}]


paramgrid = dict(penalty=penalty, eta0=eta0, class_weight=class_weight, loss=loss, alpha=alpha, learning_rate=learning_rate)

grid2 = GridSearchCV(estimator=sgd, param_grid=paramgrid, scoring='roc_auc', verbose=1, n_jobs=-1)
grid_result2 = grid2.fit(X_train, y_train)
```

```
grid_result2.best_score_
grid_result2.best_params_
```

```
y_test_pred2 = grid_result2.predict(X_test)
```
